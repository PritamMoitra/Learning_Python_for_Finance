Importing necessary libraries

```python
import pandas as pd
import yfinance as yf
```

Download Sock data

```python
df=yf.download(tickers=['ITC.NS','ARE&M.NS','CAPLIPOINT.NS'],period='3y',interval='1mo')
```

join multi-level columns and rename the columns

```python
df.columns=['_'.join(col).strip() for col in df.columns.values]
print(df.head)
```

save in csv format

```python
df.to_csv(path_or_buf='stock_data.csv' , decimal = '.' , sep= ',', header=True)
```

save in tsv format

```python
df.to_csv(path_or_buf='stock_data.tsv' , decimal = '.' , sep= '\t', header=True)
```

read & print the info

```python
df_csv=pd.read_csv(filepath_or_buffer='stock_data.csv' , index_col=['Date'],parse_dates=True,decimal='.',sep=',')
print(df_csv.info())
print(df_csv.head())
df_tsv=pd.read_csv(filepath_or_buffer='stock_data.tsv' , index_col=['Date'],parse_dates=True,decimal='.',sep='\t')
print(df_tsv.info())
print(df_tsv.head())
```
